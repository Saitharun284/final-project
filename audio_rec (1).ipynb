```python
def get_text_emotion(df:pd.DataFrame, text:str, error_scale=10):

    from transformers import pipeline
    classifier = pipeline("sentiment-analysis", model="michellejieli/emotion_text_classifier")
    model_output = classifier(text)[0]

    category = model_output['label']

    prompt = f"Recommend some activities to do for a person in this particular {category}"


    completion = client.chat.completions.create(
      model="gpt-3.5-turbo",
      messages=[
        {"role": "system", "content": "You are a doctor who gives psychological advices to person"},
        {"role": "user", "content": prompt}
      ]
    )
    score = model_output['score']

    max_tempo = df.tempo.max()
    tempo_calculation = score * max_tempo

    cmap = mapping_dict[category]
    filtered_df = df[df.category == cmap].reset_index(drop=True)

    low_range = tempo_calculation - error_scale
    upper_range = tempo_calculation + error_scale

    tempo_match = filtered_df.tempo.between(low_range, upper_range, inclusive='both')
    recommended = filtered_df[tempo_match].filepath.values.tolist()


    return recommended[:3],completion.choices[0].message
```

## Music Files Tempo Generation

```python
import glob

music_files = glob.glob("downloads_music/*/*.mp3")
print(len(music_files))
```

```python
from tqdm import tqdm
import librosa
import pandas as pd
import numpy as np

metadata_df = pd.DataFrame(index=np.arange(len(music_files)), columns=['filepath', 'tempo', 'spec_centroid', 'zcr'])
for i in tqdm(range(len(music_files))):
    y, sr = librosa.load(music_files[i])
    tempo, beats_track = librosa.beat.beat_track(y=y, sr=sr)
    spec_centroid = np.mean(librosa.feature.spectral_centroid(y= y, sr= sr))
    zcr = np.mean(librosa.feature.zero_crossing_rate(y=y))

    metadata_df.loc[i, "filepath"] = music_files[i]
    metadata_df.loc[i, "tempo"] = tempo
    metadata_df.loc[i, "spec_centroid"] = spec_centroid
    metadata_df.loc[i, "zcr"] = zcr
```

```python
import json
test_data = [json.load(open("EmotionLines_friends_annotation/EmotionLines/Friends/friends_test.json", "r"))]
```

```python
samples = []
for x in range(len(test_data[0])):
   samples.extend(test_data[0][x])
```

```python
len(samples)
```

```python
import pandas as pd

classes_needed = ["anger", "joy", "surprise", "disgust", "fear", "sadness"]
samples_dist = pd.DataFrame([{"utterance":x['utterance'], "emotion": x['emotion']} for x in samples])
samples_dist = samples_dist[samples_dist.emotion.isin(classes_needed)].reset_index(drop=True)
```

```python
samples_dist.emotion.value_counts()
```

```python
emotion_model = AutoModel.from_pretrained("michellejieli/emotion_text_classifier")
model_tokenizer = AutoTokenizer.from_pretrained("michellejieli/emotion_text_classifier")
```

```python
from transformers import pipeline
from tqdm import tqdm

samples_dist['model_pred'] = ""
for i in tqdm(range(len(samples_dist))):

    text = samples_dist.loc[i, "utterance"]

    classifier = pipeline("sentiment-analysis", model="michellejieli/emotion_text_classifier")
    model_output = classifier(text)[0]

    category = model_output['label']
    score = model_output['score']

    samples_dist.loc[i, "model_pred"] = category
```

```python
import pandas as pd
from sklearn.metrics import recall_score

metrics_data = pd.read_csv("metrics.csv")
classes_needed = ["anger", "joy", "surprise", "disgust", "fear", "sadness"]

for cl in classes_needed:
    df = metrics_data[metrics_data.emotion == cl].reset_index(drop=True)
    recall = recall_score(y_true=df['emotion'], y_pred=df['model_pred'], average="weighted")

    print(f"Emotion: {cl} Num Samples: {len(df)} Recall score: {recall}")
```
